```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression # LR is Class

# Sample Data (you can replace it with your own data)
data = {
    'X': [100,200,300, 400, 500, 750, 1000],  # Feature (e.g., house size, years of experience)
    'Y': [ 2,  3,  5,   7,  11,   16,  20 ]  # Target (e.g., house price, salary)
}

# Load data into a pandas DataFrame
df = pd.DataFrame(data)

# Feature and target arrays
X = df[['X']].values  # Feature should be 2D (Scikit-learn expects this format)
Y = df['Y'].values    # Target is 1D
print(type(X))
print(X.shape)
print(type(Y))
# Create and train the linear regression model
model = LinearRegression()
model.fit(X, Y)

# Extract the slope and intercept from the trained model
slope = model.coef_[0]
intercept = model.intercept_

# Print the calculated parameters
print(f"Intercept (b): {intercept}")
print(f"Slope (m): {slope}")

# Make predictions using the model
Y_pred = model.predict(X)
print(type(Y_pred))
print(Y_pred.shape)

# Plot the data and the regression line
plt.scatter(X, Y, color='blue', label='Data Points')
plt.plot(X, Y_pred, color='red', label='Best Fit Line')
plt.xlabel('Size in Sq Ft')
plt.ylabel('Plot Price in Million')
plt.title('Linear Regression')
plt.legend()
plt.show()
```
